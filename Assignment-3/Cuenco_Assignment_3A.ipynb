````python
# extracting dirty list of courses 

import csv

dirty_course_list = list()

with open('recweekdata.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            pass
        else:
            dirty_course_list.append(row[5])
        line_count += 1
        
dirty_course_list
````

````python
# identifying unique course names

unique_courses = set(dirty_course_list)

unique_courses
````

````python
# making course dictionary

unique_courses_dictionary = {course:None for course in unique_courses}

unique_courses_dictionary
````

````python
# changing dirty course names to proper/clean course names

for i in unique_courses_dictionary:
    final_course = input(i + ": " + "Enter name: ")
    unique_courses_dictionary[i] = final_course
````

````python
# creating new list for courses with clean course list

import re

clean_course_list = [re.sub(r''.join(course),r''.join(unique_courses_dictionary[course]),course) for course in dirty_course_list]

clean_course_list
````

````python
# extracting dirty list of years

import csv

dirty_year_list = list()

with open('recweekdata.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            pass
        else:
            dirty_year_list.append(row[4])
        line_count += 1
        
dirty_year_list
````

````python
clean_year_list = []

for year in dirty_year_list:
    if len(year) == 2: # example: 97, 98, 99
        clean_year_list.append(year)
    else: # example: 1997, 1998, 1999
        newyear = year[2:]
        clean_year_list.append(newyear)
        
clean_year_list
````

````python
# extracting dirty list of months

import csv

dirty_month_list = list()

with open('recweekdata.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            pass
        else:
            dirty_month_list.append(row[2])
        line_count += 1
        
dirty_month_list
````

````python
# identifying unique month names

unique_month = set(dirty_month_list)

unique_month
````

````python
# making month dictionary

unique_month_dictionary = {month:None for month in unique_month}

unique_month_dictionary
````

````python
numbermonth_dict = {'1':'January','2':'February','3':'March','4':'April','5':'May','6':'June','7':'July','8':'August','9':'September','10':'October','11':'November','12':'December'}

numbermonth_dict
````

````python
shortmonth_dict = {'Jan':'January','Feb':'February','Mar':'March','Apr':'April','May':'May','Jun':'June','Jul':'July','Aug':'August','Sep':'September','Oct':'October','Nov':'November','Dec':'December'}

shortmonth_dict
````

````python
# changing dirty month names to proper/clean month names

for i in unique_month_dictionary:
    if len(i) == 3: # example: Jan, Feb, Mar
        final_month = shortmonth_dict[i]
    elif len(i) < 3: # example: 1, 2, 3
        final_month = numbermonth_dict[i]
    else:
        final_month = i
    unique_month_dictionary[i] = final_month
````

````python
# creating new list for months with clean month list

clean_month_list = [re.sub(r''.join(month),r''.join(unique_month_dictionary[month]),month) for month in dirty_month_list]

clean_month_list
````

````python
# extracting dirty list of phone numbers

import csv

dirty_phone_list = list()

with open('recweekdata.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            pass
        else:
            dirty_phone_list.append(row[6])
        line_count += 1
        
dirty_phone_list
````

````python
# changing dirty phone number to proper/clean phone number

clean_phone_list = []

for number in dirty_phone_list:
    if len(number) == 16: # example: +63 999 999 9999 = len of example is 16
        clean_phone_list.append(number)
    else: # example: 999 999 9999
        newnumber = "+63 " + number # add "+63 " at the start of the number to have a similar format as previous example
        clean_phone_list.append(newnumber)
        
clean_phone_list
````

````python
# creating new list for phone number with clean phone number list

clean_number_list = []

for number in clean_phone_list:
    new = number.replace(" ","") # removes spaces
    clean_number_list.append(new)
    
clean_number_list
````

````python
# extracting first name list

import csv

firstname_list = list()

with open('recweekdata.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            pass
        else:
            firstname_list.append(row[0])
        line_count += 1
        
firstname_list
````

````python
# extracting last name list

import csv

lastname_list = list()

with open('recweekdata.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            pass
        else:
            lastname_list.append(row[1])
        line_count += 1
        
lastname_list
````

````python
# extracting day list

import csv

day_list = list()

with open('recweekdata.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            pass
        else:
            day_list.append(row[3])
        line_count += 1
        
day_list
````

````python
# this is what you will see in the new clean csv file
# practice

print('firstname,'+'lastname,'+'birthmonth,'+'birthday,'+'birthyear,'+'course,'+'phonenumber')
for i in range(len(firstname_list)+1):
    print(firstname_list[i]+","+lastname_list[i]+","+clean_month_list[i]+","+day_list[i]+","+clean_year_list[i]+","+clean_course_list[i]+","+clean_number_list[i])
````

````python
# final
# use zip to concatenate corresponding information (first name, last name, birth month, birth day, birth year, course, phone number) for each person
# will end up with the same format as the practice above

rows = zip(firstname_list,lastname_list,clean_month_list,day_list,clean_year_list,clean_course_list,clean_number_list)
````

````python
# creating a new csv file containing clean list

import csv

with open('cleanrecweekdata.csv', 'a', newline='') as csvfile:
    order_writer = csv.writer(csvfile, quoting=csv.QUOTE_MINIMAL)
    order_writer.writerow(["firstname","lastname","birthmonth","birthday","birthyear","course","phonenumber"])
    for row in rows:
        order_writer.writerow(row)
````
