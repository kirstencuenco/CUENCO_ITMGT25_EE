````python
## Code goes here

import pandas as pd

rwga = pd.read_csv("recweekdata.csv")

print(rwga)
````

# DATA CLEANSING

## MONTH

````python
rwga["birthmonth"] = rwga["birthmonth"].str.replace("^1$","January")
rwga["birthmonth"] = rwga["birthmonth"].str.replace("^Jan$","January")

rwga["birthmonth"] = rwga["birthmonth"].str.replace("^2$","February")
rwga["birthmonth"] = rwga["birthmonth"].str.replace("^Feb$","February")

rwga["birthmonth"] = rwga["birthmonth"].str.replace("^3$","March")
rwga["birthmonth"] = rwga["birthmonth"].str.replace("^Mar$","March")

rwga["birthmonth"] = rwga["birthmonth"].str.replace("^4$","April")
rwga["birthmonth"] = rwga["birthmonth"].str.replace("^Apr$","April")

rwga["birthmonth"] = rwga["birthmonth"].str.replace("^5$","May")

rwga["birthmonth"] = rwga["birthmonth"].str.replace("^6$","June")
rwga["birthmonth"] = rwga["birthmonth"].str.replace("^Jun$","June")

rwga["birthmonth"] = rwga["birthmonth"].str.replace("^7$","July")
rwga["birthmonth"] = rwga["birthmonth"].str.replace("^Jul$","July")

rwga["birthmonth"] = rwga["birthmonth"].str.replace("^8$","August")
rwga["birthmonth"] = rwga["birthmonth"].str.replace("^Aug$","August")

rwga["birthmonth"] = rwga["birthmonth"].str.replace("^9$","September")
rwga["birthmonth"] = rwga["birthmonth"].str.replace("^Sep$","September")

rwga["birthmonth"] = rwga["birthmonth"].str.replace("^10$","October")
rwga["birthmonth"] = rwga["birthmonth"].str.replace("^Oct$","October")

rwga["birthmonth"] = rwga["birthmonth"].str.replace("^11$","November")
rwga["birthmonth"] = rwga["birthmonth"].str.replace("^Nov$","November")
rwga["birthmonth"] = rwga["birthmonth"].str.replace("^12$","December")
rwga["birthmonth"] = rwga["birthmonth"].str.replace("^Dec$","December")

rwga["birthmonth"]
````

## UPDATED RECWEEK DATA WITH CLEAN MONTH

````python
rwga
````

## YEAR

````python
rwga["birthyear"] = rwga.birthyear.astype(str)

rwga["birthyear"] = rwga["birthyear"].str.replace("^1997$","97")
rwga["birthyear"] = rwga["birthyear"].str.replace("^1998$","98")
rwga["birthyear"] = rwga["birthyear"].str.replace("^1999$","99")
rwga["birthyear"] = rwga["birthyear"].str.replace("^2000$","00")
rwga["birthyear"] = rwga["birthyear"].str.replace("^2001$","01")

rwga["birthyear"]
````

## UPDATED RECWEEK DATA WITH CLEAN YEAR

````python
rwga
````

## COURSE

````python
rwga["course"] = rwga["course"].str.replace("^BS LM$","BS Legal Management")
rwga["course"] = rwga["course"].str.replace("^LM$","BS Legal Management")
rwga["course"] = rwga["course"].str.replace("^Legal Management$","BS Legal Management")

rwga["course"] = rwga["course"].str.replace("^BS CTM$","BS Communications Technology Management")
rwga["course"] = rwga["course"].str.replace("^CTM$","BS Communications Technology Management")
rwga["course"] = rwga["course"].str.replace("^Communications Technology$","BS Communications Technology Management")

rwga["course"] = rwga["course"].str.replace("^BS ME$","BS Management Engineering")
rwga["course"] = rwga["course"].str.replace("^ME$","BS Management Engineering")
rwga["course"] = rwga["course"].str.replace("^Management Engineering$","BS Management Engineering")

rwga["course"] = rwga["course"].str.replace("^BS MGT$","BS Management")
rwga["course"] = rwga["course"].str.replace("^MGT$","BS Management")
rwga["course"] = rwga["course"].str.replace("^Management$","BS Management")

rwga["course"] = rwga["course"].str.replace("^BS ITE$","BS Information Technology Entrepreneurship")
rwga["course"] = rwga["course"].str.replace("^ITE$","BS Information Technology Entrepreneurship")
rwga["course"] = rwga["course"].str.replace("^Information Technology Entrepreneurship$","BS Information Technology Entrepreneurship")

rwga["course"]
````

## UPDATED RECWEEK DATA WITH CLEAN COURSE

````python
rwga
````

## PHONENUMBER

````python
rwga["phonenumber"] = rwga["phonenumber"].str.replace("^9","+639")
rwga["phonenumber"] = rwga["phonenumber"].str.replace(" ","")

rwga["phonenumber"]
````

## UPDATED RECWEEK DATA WITH CLEAN PHONENUMBER

````python
rwga
````

# MORE IMPORTANT DATA

## IDENTIFYING CARRIER

````python
carrier = {"+63917":"Globe","+63927":"Globe","+63918":"Smart"}

rwga["carrier"] = rwga.loc[:,"phonenumber"].str[0:6].apply(lambda x:carrier[x])

rwga
````

## GENDER LIST

````python
profile = pd.read_csv("studentprofiles.csv")
profile = profile.drop(columns=["lastname"])

print(profile)
````

## RECWEEK DATA MERGED WITH GENDER

````python
merged = pd.merge(rwga,profile,left_on=["firstname"],right_on="firstname")

merged
````
